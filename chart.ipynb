```python
import json

def generate_html(tasks):
    html_template = '''
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <title>Gantt Chart</title>
        <script src="https://www.gstatic.com/charts/loader.js"></script>
        <style>
            .container {{
                max-width: 1200px;
                margin: 0 auto;
                padding: 20px;
            }}
            .chart {{
                height: 600px;
            }}
        </style>
    </head>
    <body>
        <div class="container">
            <h1>Gantt Chart</h1>
            <div id="chart_div" class="chart"></div>
        </div>
        <script>
            const tasks = {tasks_json};

            google.charts.load('current', {{packages: ['gantt']}});
            google.charts.setOnLoadCallback(drawChart);

            function parseDate(dateString) {{
                var parts = dateString.split(/[-T:Z]/);
                return new Date(Date.UTC(parts[0], parts[1] - 1, parts[2], parts[3] || 0, parts[4] || 0, parts[5] || 0));
            }}

            function drawChart() {{
                var data = new google.visualization.DataTable();
                data.addColumn('string', 'Task ID');
                data.addColumn('string', 'Task Name');
                data.addColumn('date', 'Start Date');
                data.addColumn('date', 'End Date');
                data.addColumn('number', 'Duration');
                data.addColumn('number', 'Percent Complete');
                data.addColumn('string', 'Dependencies');

                {tasks_rows}

                var options = {{
                    height: 600,
                    gantt: {{
                        trackHeight: 30,
                        criticalPathEnabled: true, // Draws attention to the critical path
                        criticalPathStyle: {{
                            stroke: '#e64a19', // Red color for critical path
                            strokeWidth: 2
                        }},
                        labelStyle: {{
                            fontSize: 12, // Ensures readability
                        }},
                        barCornerRadius: 4,
                        arrow: {{
                            angle: 45,
                            width: 1.5,
                            color: 'black',
                            radius: 0
                        }}
                    }},
                    hAxis: {{
                        format: 'MMM dd, yyyy HH:mm',
                        gridlines: {{
                            color: '#a6a6a6', // Darker color for horizontal gridlines
                            count: 10 // Increase gridline count for better visibility
                        }},
                        minorGridlines: {{
                            color: '#e0e0e0', // Lighter color for minor horizontal gridlines
                            count: 4
                        }},
                        textStyle: {{
                            fontSize: 12
                        }},
                    }},
                    vAxis: {{
                        gridlines: {{
                            color: '#a6a6a6' // Darker color for vertical gridlines
                        }},
                        textStyle: {{
                            fontSize: 12
                        }}
                    }},
                    backgroundColor: '#ffffff'
                }};

                var chart = new google.visualization.Gantt(document.getElementById('chart_div'));
                chart.draw(data, options);
            }}
        </script>
    </body>
    </html>
    '''

    tasks_rows = ""
    for task in tasks:
        task_id = task['id']
        task_name = task['name']
        start_date = task['start_date']
        end_date = task['end_date']

        # As duration, percent_complete, and dependencies are not provided in your data, they default to 'null'
        tasks_rows += (
            f"data.addRows([['{task_id}', '{task_name}', "
            f"parseDate('{start_date}'), parseDate('{end_date}'), "
            f"null, null, null]]);\n"
        )

    tasks_json = json.dumps(tasks)

    return html_template.format(tasks_rows=tasks_rows, tasks_json=tasks_json)

```
